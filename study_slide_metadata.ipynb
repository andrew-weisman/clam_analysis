```python
areas_duped = df.area_sq_px.loc[df.area_sq_px.duplicated()].to_list()
duped_images = df.loc[df.area_sq_px.apply(lambda x: x in areas_duped)]
```

Note, from the above `duped_images` dataframe, even though `DigitalSlide_B1M_18S_1.mrxs` and `DigitalSlide_B2M_4S_1_20210826184014839.mrxs` have the exact same sizes and other properties in common, they are completely different images, as confirmed from looking at the images themselves in Case Viewer.

However, the other two possible duplicates found, `DigitalSlide_B2M_1S_1.mrxs` and `DigitalSlide_B2M_1S_1_20210826183728179.mrxs`, are indeed duplicates. This can be determined from the following code:
