Implement a naive check that the third zoom level in all the images is not present in the other zooms to within a power of 2.

````python
# Import numpy and print out the unique true magnification/power/zoom levels
import numpy as np
unique_resolutions = df.output_resolution.unique()
print(unique_resolutions)
````

````python
# Multiply the zoom levels (converted to microns per pixel) by num_powers different powers of two
num_powers = 20
mpp_all = np.tile(2**(np.arange(num_powers)-num_powers/2)[:, None], (1, 3)) * np.tile((10 / unique_resolutions)[None, :], (num_powers, 1))
print(mpp_all)
````

````python
# Print the number of times each zoom level is equivalent to the other to within a power of two
# This shows that the first two zoom levels are equivalent to each other to within a single power of two, whereas the last is not equivalent to either
tol = 1e-8
print((np.abs(mpp_all[:, 0][:, None] - mpp_all[:, 0][None, :]) < tol).sum())
print((np.abs(mpp_all[:, 0][:, None] - mpp_all[:, 1][None, :]) < tol).sum())
print((np.abs(mpp_all[:, 0][:, None] - mpp_all[:, 2][None, :]) < tol).sum())
print((np.abs(mpp_all[:, 1][:, None] - mpp_all[:, 0][None, :]) < tol).sum())
print((np.abs(mpp_all[:, 1][:, None] - mpp_all[:, 1][None, :]) < tol).sum())
print((np.abs(mpp_all[:, 1][:, None] - mpp_all[:, 2][None, :]) < tol).sum())
print((np.abs(mpp_all[:, 2][:, None] - mpp_all[:, 0][None, :]) < tol).sum())
print((np.abs(mpp_all[:, 2][:, None] - mpp_all[:, 1][None, :]) < tol).sum())
print((np.abs(mpp_all[:, 2][:, None] - mpp_all[:, 2][None, :]) < tol).sum())
````

Add columns to the dataframe containing appropriate downsampling-related parameters

````python
# INPUT
further_reduction_fac_on_smallest_resolution = 2

# Import relevant library
import numpy as np

# Parameter for determining whether the log2 of the downsample factor is an integer
tol = 1e-6

# Common resolution we want to make every image
target_resolution = unique_resolutions.min() / further_reduction_fac_on_smallest_resolution
target_patch_size = 256

# Downsample factors to achieve the target resolution
downsample_factors = unique_resolutions / target_resolution

# log2 of the downsample factors
power_of_2_factors = np.log2(downsample_factors)

# Whether the log2 of the downsample factors are integers to within the desired tolerance
are_power_of_2_factors = np.abs(power_of_2_factors - power_of_2_factors.round()) < tol

# Patch sizes, assuming this needs to be an integer
patch_sizes = (downsample_factors * target_patch_size).round().astype('int')

# Output for each resolution present in the dataset
print('Parameters needed to achieve {} resolution for all images:'.format(target_resolution))
mapper_level = {}
mapper_patch_size = {}
mapper_downsample_factor = {}
for unique_resolution, downsample_factor, patch_size, power_of_2_factor, is_power_of_2_factor in zip(unique_resolutions, downsample_factors, patch_sizes, power_of_2_factors, are_power_of_2_factors):
    print('  ----')
    print('  unique_resolution: {}'.format(unique_resolution))
    if is_power_of_2_factor:
        level = np.round(power_of_2_factor).astype('int')
        patch_size = target_patch_size
        # print('  level: {:d}'.format(level))
        # downsample_factor = None
    else:
        # level = None
        level = 0
        patch_size = patch_size
        # print('  downsample_factor: {:0.4f}'.format(downsample_factor))
    downsample_factor = None
    print('  level: {}'.format(level))
    print('  patch_size: {}'.format(patch_size))
    mapper_level[unique_resolution] = level
    mapper_patch_size[unique_resolution] = patch_size
    mapper_downsample_factor[unique_resolution] = downsample_factor

# Add relevant columns to the dataframe
df['level'] = df.output_resolution.map(mapper_level)
df['patch_size'] = df.output_resolution.map(mapper_patch_size)
df['downsample_factor'] = df.output_resolution.map(mapper_downsample_factor)
````

Place different-sized images into different process lists and print the patching and feature extraction commands to be run on them.

```bash

```

````python
# INPUT
preset = 'bwh_resection'

# Import relevant library
import numpy as np
from datetime import datetime

# Add to the dataframe a process_list column containing the name of the process list file each row should correspond to
df['process_list'] = df.apply(lambda x: 'filenames_of_{:d}x.csv'.format(int(np.round(x['output_resolution']))), axis='columns')

# Get a pretty string versions needed for below
target_resolution_pretty = '{:d}x'.format(int(np.round(target_resolution)))
datetime_pretty = datetime.today().strftime('%Y-%m-%d-%H%M%S')

# Add a column containing the patching command the row should correspond to
# df['patching_command'] = df.apply(lambda x: 'create_patches_fp.py --process_list {} --patch_level {:d} --patch_size {:d} --step_size {:d} --preset MY-PRESET --save_dir /path/to/all/patches'.format(x['process_list'], int(x['level']), x['patch_size'], x['patch_size']), axis='columns')
df['patching_command'] = df.apply(lambda x: 'python $CLAM/create_patches_fp.py --process_list $working_dir/{} --patch_level {:d} --patch_size {:d} --step_size {:d} --preset $CLAM/presets/{}.csv --source $working_dir/data --save_dir $working_dir/results/{}/bwh_resection --seg --patch --stitch 2>&1 | tee -a $working_dir/logs/preprocessing-{}.log'.format(x['process_list'], int(x['level']), x['patch_size'], x['patch_size'], preset, target_resolution_pretty, datetime_pretty), axis='columns')

# Print the commands that should be run
print('Patching command(s) that should be run:')
print(df['patching_command'].unique())
print('')
print('Feature extraction command that should be run:')
# print('extract_features.py --target_patch_size {:d} --data_h5_dir /path/to/all/patches --data_slide_dir /path/to/all/images --csv_path /path/to/all/images.csv'.format(target_patch_size))
print('python $CLAM/extract_features_fp.py --target_patch_size {:d} --data_h5_dir $working_dir/results/{}/bwh_resection --data_slide_dir $working_dir/data --csv_path $working_dir/feature_extraction_input.csv --feat_dir $working_dir/results/{}/bwh_resection/features --batch_size 895 --slide_ext .mrxs 2>&1 | tee -a $working_dir/logs/feature_extraction-{}.log'.format(target_patch_size, target_resolution_pretty, target_resolution_pretty, datetime_pretty))

# Read in the preset file, dropping the unnecessary columns for the process list
df_preset = pd.read_csv(filepath_or_buffer='/home/weismanal/projects/idibell/links/clam_installation/repo/presets/{}.csv'.format(preset)).drop(['white_thresh', 'black_thresh'], axis='columns')

# Add three columns to the preset that are needed for the process list
df_preset['status'] = None
df_preset['process'] = 1
df_preset['slide_id'] = 'XXXX.mrxs'

# Get a series version of this template dataframe
template_series = df_preset.iloc[0, :]

# For each unique process list filename...
for process_list in df['process_list'].unique():

    # Get the corresponding dataframe slice
    curr_df = df[df['process_list'] == process_list]

    # Initialize the list that will hold the series that need to be concatenated to a full dataframe and then written to disk
    row_holder = []

    # For each basename filename in the current dataframe...
    for slide_id in curr_df['image_pathname'].apply(lambda x: x.split('/')[-1]):

        # Copy the template series
        curr_row = template_series.copy()

        # In the copy, set the slide_id needed for the process list
        curr_row['slide_id'] = slide_id

        # Add the copy of the template series to the list holding all the series to be concatenated
        row_holder.append(curr_row)

        # Concatenate the series into a dataframe, set the index to the slide_id, and write the dataframe to disk as the current process list
        pd.concat(row_holder, axis='columns').T.set_index('slide_id').to_csv(path_or_buf=process_list)
````
